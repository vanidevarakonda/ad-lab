```python
import spacy
nlp = spacy.load('en_core_web_sm')
doc = nlp(u'Apple is ready to launch new phone worth $1000 in New york time square ')
for ent in doc.ents:
    print(ent.text, ent.start_char, ent.end_char, ent.label_)
```

##USE CASE

```python
import spacy
nlp = spacy.load("en_core_web_sm")
```

```python
text = """
Elon Musk is a billionaire entrepreneur, engineer, and inventor known for his role in revolutionizing multiple industries, including electric vehicles, space exploration, and artificial intelligence. Born in 1971 in Pretoria, South Africa, he co-founded companies such as PayPal, Tesla, SpaceX, Neuralink, and The Boring Company."""
```

```python
doc = nlp(text)

for ent in doc.ents:
    print(f"Entity: {ent.text}, Label: {ent.label_}")
```

```python
from spacy import displacy
displacy.render(doc, style="ent")
```

```python
import pandas as pd
entities = [(ent.text, ent.label_, ent.lemma_) for ent in doc.ents]
df = pd.DataFrame(entities, columns=['text','type','lemma'])
print(df)
```
